```python
sales_data = np.array([
    [1,120, 135, 150, 165, 180],  # Restaurant 1
    [2,90,  100, 110, 130, 145],  # Restaurant 2
    [3,200, 210, 220, 230, 250],  # Restaurant 3
    [4,60,  75,  90,  105, 120],  # Restaurant 4
    [5,150, 160, 170, 185, 200]   # Restaurant 5
])
print(sales_data.shape)
```

```python
print(sales_data[0:3,1:6])
```

```python
print(np.sum(sales_data,axis=0))
```

```python
print(np.sum(sales_data,axis=1))
```

```python
print(np.min(sales_data[:,1:],axis=1))
```

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
avg_sales=np.mean(sales_data[:,1:],axis=1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales across all restraunt")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
